```python
import pandas as pd
```

```python
df = pd.read_csv('data/1januarynew.csv', header=0, sep=',')
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df = df.reset_index(drop=True)
```

```python
df.head()
```

```python
pickups = df.loc[: , ['pickup_datetime','pickup_neighborhood'] ]
```

```python
pickups.head()
```

```python
pickups['dummy'] = 1
```

```python
pickups.head()
```

```python
pickups_groups = pickups.groupby(['pickup_neighborhood'])
```

```python
sum_ = pickups_groups.sum()
sum_.sort_values('dummy', inplace=True, ascending=False)
```

```python
sum_
```

```python
import matplotlib.pyplot
```

```python
pickup_plot = sum_.plot(kind='bar', figsize = (15,12))
```

```python
pickup_fig = pickup_plot.get_figure()
```

```python
pickup_fig
```

```python
pickup_fig.savefig('pickups')
```

Let's see how many dropoffs we have for each neighborhood

```python
dropoffs = df.loc[: , ['dropoff_datetime','dropoff_neighborhood'] ]
dropoffs['dummy'] = 1
dropoffs_groups = dropoffs.groupby(['dropoff_neighborhood'])
```

```python
sum_2 = dropoffs_groups.sum()
sum_2.sort_values('dummy',inplace=True, ascending=False)
sum_2
```

```python
dropoff_plot = sum_2.plot(kind='bar', figsize = (15,12))
dropoff_fig = dropoff_plot.get_figure()
dropoff_fig
```

```python
dropoff_fig.savefig('dropoffs')
```
