```python
Coord_Translated = pd.read_csv('data/MN3.csv')
Coord_Translated.drop(['LandUse','AssessTot'], axis=1, inplace=True)
Coord_Translated
```

```python
PLUTO_MN = PLUTO_MN.merge(Coord_Translated, on ="Address", suffixes= ('_NY_LI','_Real'))
```

```python
PLUTO_MN.columns
```

```python
PLUTO_MN['place'] = -1
R = 6378137
cellSize = 1000
south_lat = 40.680396
weast_lon = -74.047285
```

```python
import math
```

```python
def setPick(x):
    stepLatitude =   cellSize *1.1* 180/(math.pi * R)
    stepLongitude =  cellSize *0.9*180 /( R* math.cos(math.pi*south_lat/180) * math.pi )
    lat =  x['XCoord_Real']
    lon = x['YCoord_Real']
    i = int(math.trunc((lat - south_lat) / stepLatitude))
    j = int(math.trunc((lon - weast_lon) / stepLongitude))
    return '{},{}'.format(i,j)
```

```python
PLUTO_MN['place'] = PLUTO_MN.apply(setPick, axis=1)
```
