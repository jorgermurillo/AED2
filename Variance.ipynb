```python
import pandas as pd
```

```python
df = pd.read_csv("data/january_manhattan2.csv")
df
```

```python
import dateutil
df['time']=df['pickup_datetime'].apply(dateutil.parser.parse)
```

```python
df.head()
```

```python
df['pick_hour']=df['time'].dt.hour
```

```python
def WeekGroup(x):
    if x==0 or x==1or x==2 or x == 3:
        return 0
    if x==4:
        return 1
    if x ==5:
        return 2
    if x==6:
        return 3
```

```python
df['pick_weekgroup']= df['pick_weekday'].apply(WeekGroup)
```
