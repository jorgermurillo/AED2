```python
import pandas as pd
```

```python
df = None
grid_pickups = pd.read_csv('grid_pickups_jan.csv')
#df = df.set_index(['pickups_place','pick_weekgroup'])
grid_pickups
```

```python
def visualizeWeekDayrides(x, grid ,filename, y_upper):
    #x = gridpick_ups
    x = grid_pickups.loc[ grid_pickups['pickups_place']==grid, ['pick_hour','mean_per_hour','pick_weekgroup'] ]
    x = x.set_index(['pick_weekgroup','pick_hour'])
    x = x.unstack(level=0)
    x
    plot1 = x.plot(figsize = (20,10) , ylim=(0,y_upper))
    fig1 = plot1.get_figure()
    img1 = fig1.savefig(filename)
```

```python
visualizeWeekDayrides(df,'16,11','16,11.png')
visualizeWeekDayrides(df,'9,6','9,6.png')
```
