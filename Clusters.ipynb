```python
import pandas as pd
import dateutil
```

```python
df = pd.read_csv('data/january_manhattan2.csv')
```

```python
df['time'] = df['pickup_datetime'].apply(dateutil.parser.parse)
df
```

```python
df['pick_hour']=df['time'].dt.hour
df
```

```python
def WeekGroup(x):
    if x==0 or x==1or x==2 or x == 3:
        return 0
    if x==4:
        return 1
    if x ==5:
        return 2
    if x==6:
        return 3
```

```python
df['weekgroup']=df['pick_weekday'].apply(WeekGroup)
df
```

```python
grid4_5 = df.loc[(df['pickups_place']=='4,5')&((df['pick_hour']<=3 )|(df['pick_hour']>=20))&(df['weekgroup']==2),:]
grid4_5
```

```python
plot4_5 = grid4_5.plot(kind='scatter', x='pickup_longitude', y='pickup_latitude', figsize=(20,20))
fig4_5 = plot4_5.get_figure()
fig4_5.savefig('4_5scatter.png')
```

```python
grid4_5['pickups_place'].value_counts()
```

```python
grid4_5_2 = df.loc[(df['pickups_place']=='4,5')&(df['pick_hour']==1 )&(df['weekgroup']==2),:]
grid4_5_2
```

```python
plot4_5_2 = grid4_5_2.plot(kind='scatter', x='pickup_longitude', y='pickup_latitude', figsize=(20,20))
fig4_5_2 = plot4_5_2.get_figure()
fig4_5_2.savefig('4_5scatter2.png')
```

```python
from sklearn.cluster import DBSCAN
```

```python
matrix = grid4_5_2.loc[:,['pickup_longitude','pickup_latitude']].as_matrix()
matrix
```

```python
db = DBSCAN(eps=0.0001, min_samples=10).fit(matrix)
```

```python
type(db.labels_)
```

```python
db.labels_.size
```

```python
db.labels_[db.labels_!=0]
```

```python
import numpy as np
np.unique(db.labels_)
```

```python
grid4_5_2['cluster']=db.labels_
```

```python
grid4_5_2.head(15)
```

```python
cluster_plot = grid4_5_2.loc[:,['pickup_longitude','pickup_latitude','cluster']].plot.scatter(x='pickup_longitude',y='pickup_latitude',c='cluster', figsize=(30,20))
```

```python
cluster_fig = cluster_plot.get_figure()
cluster_fig.savefig('cluster4_5.png')
```
