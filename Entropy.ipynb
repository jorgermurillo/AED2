```python
def Entropy(data, cuadrant):
    grid = data.loc[ (data['pickups_place']==cuadrant) ,['dropoff_place','dummy']]    
    # Get the total amount of rides (not unique) in that time period
    totalrides = grid['dummy'].sum()
    # Make a DataFrame called entropy that has the freq of each dropof destination
    entropy = pd.DataFrame(grid['dropoff_place'].value_counts())
    entropy['freq']  =entropy['dropoff_place']
    entropy.drop('dropoff_place', axis=1, inplace=True)
    #calculate probability as freq(x)/Totalrides
    entropy['prob'] = entropy['freq']/totalrides
    import math
    # Calculate log(prob)
    entropy['log_prob'] = entropy['prob'].apply(lambda x:math.log(x,2))
    # Calculate the product of prob *loc(prob)
    entropy['product'] = entropy['prob']*entropy['log_prob']
    # Sum of products times -1 is the entropy value
    Entropy = -1*entropy['product'].sum()
    return Entropy

```

```python
def SpatialDiversity(data, cuadrant):
    grid = data.loc[ (data['pickups_place']==cuadrant) ,['dropoff_place','dummy']]    
    # Get the total amount of rides (not unique) in that time period
    totalrides = grid['dummy'].sum()
    #
    unique_dropoff_places = grid['dropoff_place'].unique().size
    #print(unique_dropoff_places)
    # Make a DataFrame called entropy that has the freq of each dropof destination
    entropy = pd.DataFrame(grid['dropoff_place'].value_counts())
    entropy['freq']  =entropy['dropoff_place']
    entropy.drop('dropoff_place', axis=1, inplace=True)
    #calculate probability as freq(x)/Totalrides
    entropy['prob'] = entropy['freq']/totalrides
    import math
    # Calculate log(prob)
    entropy['log_prob'] = entropy['prob'].apply(lambda x:math.log(x,2))
    # Calculate the product of prob *loc(prob)
    entropy['product'] = entropy['prob']*entropy['log_prob']
    # Sum of products times -1 is the entropy value
    Entropy = -1*entropy['product'].sum()
    return Entropy/math.log(unique_dropoff_places,2)
```

```python
def SpatialDiversityIncoming(data, cuadrant):
    grid = data.loc[ (data['dropoff_place']==cuadrant) ,['pickups_place','dummy']]    
    # Get the total amount of rides (not unique) in that time period
    totalrides = grid['dummy'].sum()
    #
    unique_pickup_places = grid['pickups_place'].unique().size
       
    #print("cuadrant: " + cuadrant)
    #print(unique_pickup_places)
    if unique_pickup_places==0:
        return "NaN"
    # Make a DataFrame called entropy that has the freq of each pickup source
    entropy = pd.DataFrame(grid['pickups_place'].value_counts())
    entropy['freq']  =entropy['pickups_place']
    entropy.drop('pickups_place', axis=1, inplace=True)
    #calculate probability as freq(x)/Totalrides
    entropy['prob'] = entropy['freq']/totalrides
    import math
    # Calculate log(prob)
    entropy['log_prob'] = entropy['prob'].apply(lambda x:math.log(x,2))
    # Calculate the product of prob *loc(prob)
    entropy['product'] = entropy['prob']*entropy['log_prob']
    # Sum of products times -1 is the entropy value
    Entropy = -1*entropy['product'].sum()
    return Entropy/math.log(unique_pickup_places,2)
```
