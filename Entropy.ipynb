```python
import pandas as pd
```

```python
data = pd.read_csv('data/manhattan_jan_time.csv')
```

```python
data
```

```python
grid4_5 = data.loc[ (data['pickups_place']=='4,5') ,:]
grid4_5['dropoff_place'].value_counts()
```

```python
grid4_5_2 = data.loc[ (data['pickups_place']=='4,5')&( data['weekgroup']==2 )&(data['pick_hour']==1) ,['dropoff_place','dummy']]
grid4_5_2['dropoff_place'].value_counts()
```

```python
dropoffs_4_5 = grid4_5_2['dropoff_place'].unique()
dropoffs_4_5 = None
```

```python
totalrides_4_5 = grid4_5['dummy'].sum()
totalrides_4_5
```

```python
entropy_4_5 = pd.DataFrame(grid4_5['dropoff_place'].value_counts())
entropy_4_5['freq']  =entropy_4_5['dropoff_place']
entropy_4_5.drop('dropoff_place', axis=1, inplace=True)
entropy_4_5
```

```python
entropy_4_5['prob'] = entropy_4_5['freq']/totalrides_4_5

entropy_4_5.head(5)
```

```python
import math
entropy_4_5['log_prob'] = entropy_4_5['prob'].apply(lambda x:math.log(x,2))
entropy_4_5['product'] = entropy_4_5['prob']*entropy_4_5['log_prob']
entropy_4_5.head(15)
```

```python
Entropy4_5 = -1*entropy_4_5['product'].sum()
Entropy4_5
```
