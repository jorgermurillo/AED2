```python
import pandas as pd
```

```python
std_jan = pd.read_csv('std_dev_jan.csv')
```

```python
std_jan = std_jan.set_index(['pickups_place','pick_weekgroup'])
std_jan.head(20)
```

```python
index = std_jan.index.get_level_values('pickups_place').unique()
index.size
```

```python

std_jan.loc['0,10','std_dev'].size
```

```python
diff_up = pd.DataFrame(columns=['pickups_place','std_dev_diff','std_dev_percent', 'variance_diff','variance_percent'])
diff_down = pd.DataFrame(columns=['pickups_place','std_dev_diff','std_dev_percent', 'variance_diff','variance_percent'])
```

```python
index = std_jan.index.get_level_values('pickups_place').unique()
error_cnt = 0
i_up=0
i_down=0
for i in range (index.size):
        
        try:
            std_0 = std_jan.loc[ (index[i],0) ,'std_dev' ]       
            std_2 = std_jan.loc[ (index[i],2) ,'std_dev' ]
            var_0 = std_jan.loc[ (index[i],0) ,'variance' ]       
            var_2 = std_jan.loc[ (index[i],2) ,'variance' ] 
            std_diff = (std_2 - std_0)
            var_diff = (var_2 - var_0)
            
            if std_diff >0:
                diff_up.loc[i_up]=[ index[i],std_diff,abs(std_diff/std_0)*100,var_diff, abs(var_diff/var_0)*100]
                i_up+=1
            else:
                diff_down.loc[i_down]=[ index[i],std_diff,abs(std_diff/std_0)*100,var_diff, abs(var_diff/var_0)*100]
                i_down+=1
        except KeyError:
            error_cnt+=1
```

```python
diff_up = diff_up.sort_values('std_dev_diff', ascending=False)
diff_up
```

```python
diff_down = diff_down.sort_values('std_dev_diff', ascending=True)
diff_down
```
