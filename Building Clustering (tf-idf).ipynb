```python
PLUTO_MN['place'] = -1
R = 6378137
cellSize = 1000
south_lat = 40.680396
weast_lon = -74.047285
```

```python
import math
```

```python
def setPick(x):
    stepLatitude =   cellSize *1.1* 180/(math.pi * R)
    stepLongitude =  cellSize *0.9*180 /( R* math.cos(math.pi*south_lat/180) * math.pi )
    lat =  x['XCoord_Real']
    lon = x['YCoord_Real']
    i = int(math.trunc((lat - south_lat) / stepLatitude))
    j = int(math.trunc((lon - weast_lon) / stepLongitude))
    return '{},{}'.format(i,j)
```

```python
PLUTO_MN['place'] = PLUTO_MN.apply(setPick, axis=1)
```

```python
import numpy as np
```

```python
def features(data):
    s = set(data.index.get_level_values(0))
    l = list(s)
    l.sort()
    s2 = set(data.index.get_level_values(1))
    l2 = list(s2)
    l2.sort()
    
    features = np.zeros((len(l),len(l2)))
    
    for e in l:
        for e2 in l2:
            try:
                features[ l.index(e) , l2.index(e2) ] = data.loc[e,e2]['ratio']
            except:
                features[ l.index(e) , l2.index(e2) ] = 0
    return features, l
                
    
```
